```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
data2015_df = pd.read_csv(r"C:\Users\daric\Desktop\HAPPINESS\data_2015.csv", dtype=object, low_memory=False)

data2015_df.head()
```

```python
new_2015data_df = data2015_df[['country', 'happiness_rank', 'happiness_score', 'gdp_per_capita','family', 'life_expectancy',
                                   'freedom', 'generosity','government_corr']].copy()
new_2015data_df.head()
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/World Happiness DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_2015data_df.to_sql(name='clean_2015', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from clean_2015', con=engine).head()
```
