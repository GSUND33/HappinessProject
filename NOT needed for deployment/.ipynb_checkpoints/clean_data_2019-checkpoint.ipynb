```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
data2019_df = pd.read_csv(r"C:\Users\daric\Desktop\HAPPINESS\2019.csv", dtype=object, low_memory=False)

data2019_df.head()
```

```python
new_2019data_df = data2019_df[['Country', 'Happiness Rank', 'Happiness Score', 'GDP per Capita','Social support', 'Life Expectancy',
                                   'Freedom', 'Generosity','Government Corruption']].copy()
new_2019data_df.head()
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/World Happiness DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_2019data_df.to_sql(name='clean_2019', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from clean_2019', con=engine).head()
```
